```python
#  Min-Max Scaling
def min_max_scaling(df, column_name):
    min_value = df[column_name].min()
    max_value = df[column_name].max()
    df[column_name] = (df[column_name] - min_value) / (max_value - min_value)
```

```python
min_max_scaling(data, 'open')
min_max_scaling(data, 'high')
min_max_scaling(data, 'low')
min_max_scaling(data, 'close')
min_max_scaling(data, 'volume')
min_max_scaling(data, 'ma')
min_max_scaling(data, 'rsi')
min_max_scaling(data, 'vwap')
min_max_scaling(data, '21atr')
min_max_scaling(data, 'bollinger_higher')
```

```python
#  Min-Max Scaling
def min_max_scaling(df, column_name):
    min_value = df[column_name].min()
    max_value = df[column_name].max()
    df[column_name] = (df[column_name] - min_value) / (max_value - min_value)
```

```python
min_max_scaling(data, 'open')
min_max_scaling(data, 'high')
min_max_scaling(data, 'low')
min_max_scaling(data, 'close')
min_max_scaling(data, 'volume')
min_max_scaling(data, 'ma')
min_max_scaling(data, 'rsi')
min_max_scaling(data, 'vwap')
min_max_scaling(data, '21atr')
min_max_scaling(data, 'bollinger_higher')
```

```python
df_reset['year'] = df_reset['date'].map(lambda x: pd.to_datetime(x,format='%Y-%m-%d %H:%M:%S').year)
df_reset['day'] = df_reset['date'].map(lambda x: pd.to_datetime(x,format='%Y-%m-%d %H:%M:%S').day)
```

```python
df_reset['month'] = df_reset['date'].map(lambda x: pd.to_datetime(x,format='%Y-%m-%d %H:%M:%S').month)
```

```python
df_reset[['open','high','low','close','volume','ma','rsi','vwap','21atr','bollinger_higher']].values.
```

```python
from tqdm import tqdm
```

```python
temp = {}
for n, g in tqdm(df_reset.groupby(by='year')):
    for n_, g_ in g.groupby(by='month'):
        for name, group in g_.groupby(by='day'):
            key = str(n)+"-"+str(n_)+"-"+str(name)
            temp[key] = group[['open','high','low','close','volume','ma','rsi','vwap','21atr','bollinger_higher']].values
```

```python
import pickle
pickle_file_path = '/Users/chinu/Downloads/adarnn/nifty.pkl'
with open(pickle_file_path, 'wb') as f:
    pickle.dump(temp, f)
print("Pickle file created successfully.")
```

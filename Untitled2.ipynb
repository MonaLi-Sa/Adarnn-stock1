```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from datetime import datetime, time, timedelta
```

```python
df = pd.read_pickle("/Users/chinu/Downloads/adarnn/Nifty_1.pkl")
```

```python
df
```

```python
df.reset_index(drop=False, inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df_monthly = df.groupby(df['date'].dt.to_period('M'))
min_trading_days = df_monthly.size().min()
df_filtered = df_monthly.apply(lambda x: x.head(min_trading_days)).reset_index(drop=True)
```

```python
df_filtered
```

```python
df_filtered['year'] = df_filtered['date'].map(lambda x: pd.to_datetime(x,format='%Y-%m-%d %H:%M:%S').year)
df_filtered['day'] = df_filtered['date'].map(lambda x: pd.to_datetime(x,format='%Y-%m-%d %H:%M:%S').day)
df_filtered['month'] = df_filtered['date'].map(lambda x: pd.to_datetime(x,format='%Y-%m-%d %H:%M:%S').month)
```

```python
desired_order = ['date', 'open', 'high', 'low', 'volume','ma','rsi','vwap','21atr','bollinger_higher','year','month','day','close']
df_filtered = df_filtered[desired_order]
```

```python
temp = {}
for n, g in tqdm(df_filtered.groupby(by='year')):
    for n_, g_ in g.groupby(by='month'):
        for name, group in g_.groupby(by='day'):
            key = str(n)+"-"+str(n_)+"-"+str(name)
            temp[key] = [group[['open','high','low','volume','ma','rsi','vwap','21atr','bollinger_higher']].values, group[['close']].values]
```
